```python
zip_path = "AIS.zip"     
csv_name = "AIS_Klaipeda_From20250908_To20251008.csv"   

with zipfile.ZipFile(zip_path, 'r') as z:
    with z.open(csv_name) as f:
        df = pd.read_csv(f)

df.head()
```

```python
from pyais import decode
from pyais.exceptions import InvalidNMEAMessageException

decoded_rows = []

for idx, line in enumerate(df["nmea_message"].dropna()):
    try:
        # Convert string to bytes
        raw = line.encode("utf-8")

        # Decode single AIS message
        msg = decode(raw)
        data = msg.asdict()

        decoded_rows.append({
            "mmsi": data.get("mmsi"),
            "msg_type": data.get("msg_type"),
            "latitude": data.get("lat"),
            "longitude": data.get("lon"),
            "speed_knots": data.get("speed"),
            "course_deg": data.get("course"),
            "heading": data.get("heading"),
            "ship_type": data.get("ship_type"),
            "nav_status": data.get("status"),
        })

    except InvalidNMEAMessageException:
        # Skip invalid or incomplete messages
        continue
    except Exception as e:
        print(f"Row {idx} skipped: {e}")
```

```python
from pyais import IterMessages

decoded_rows = []

ais_bytes = (
    line.encode("utf-8")
    for line in df["nmea_message"].dropna()
    if line.startswith("!AIVDM")
)

for msg in IterMessages(ais_bytes):
    data = msg.asdict()

    decoded_rows.append({
        "mmsi": data.get("mmsi"),
        "msg_type": data.get("msg_type"),
        "latitude": data.get("lat"),
        "longitude": data.get("lon"),
        "speed_knots": data.get("speed"),
        "course_deg": data.get("course"),
        "heading": data.get("heading"),
        "ship_type": data.get("ship_type"),
        "nav_status": data.get("status"),
    })
```

```python
decoded_df = pd.DataFrame(decoded_rows)
decoded_df.head()
```
