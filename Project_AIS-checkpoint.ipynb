```python
import pandas as pd
import zipfile
from pyais import IterMessages
```

```python
zip_path = "AIS.zip"     
csv_name = "AIS_Klaipeda_From20250908_To20251008.csv"   

with zipfile.ZipFile(zip_path, 'r') as z:
    with z.open(csv_name) as f:
        df = pd.read_csv(f)

df.head()
```

```python
ais_bytes = (
    line.encode("utf-8")
    for line in df["nmea_message"].dropna()
    if "!AIVDM" in line
)
```

```python
def extract_aivdm(line):
    idx = line.find("!AIVDM")
    if idx == -1:
        return None
    return line[idx:]

ais_bytes = (
    extract_aivdm(line).encode("utf-8")
    for line in df["nmea_message"].dropna()
    if extract_aivdm(line) is not None
)
```

```python
from pyais import IterMessages

count = 0

for msg in IterMessages(ais_bytes):
    print("=" * 60)
    print(msg)                 
    print(msg.asdict())        
    
    count += 1
    if count == 5:             
        break
```

```python
from pyais import decode

raw = df["nmea_message"].iloc[1].encode("utf-8")
msg = decode(raw)

print(msg)
print(msg.asdict())
```
